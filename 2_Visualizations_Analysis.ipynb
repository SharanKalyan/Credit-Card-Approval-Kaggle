### distributions and outliers

```python
def distributions():
    for i in df.columns:
        if (df[i].dtypes != 'object') & (i != 'Applicant_ID'):
            print('\n')
            print(i)
            plt.figure(figsize=(12,5))
            sns.histplot(data = df, x = i,kde=True,color='b')
            plt.grid()
            plt.title(i)
            plt.show()
            
            plt.figure(figsize=(12,5))
            sns.boxplot(data = df, x = i,color='y')
            plt.grid()
            plt.title(i)
            plt.show()
```
