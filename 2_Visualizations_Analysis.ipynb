### Lets now create a new feature that gives a genralized debt score

```python
df['Debt_Score'] =  df.Total_Good_Debt - df.Total_Bad_Debt
```

```python
col = df.pop('Status')
df.insert(23,'Status',col)
```

```python
sns.kdeplot(df.Debt_Score)
plt.grid()
plt.show()
```

```python
sns.boxplot(df.Debt_Score)
q1 = np.percentile(df['Debt_Score'], 25, interpolation='midpoint')
q3 = np.percentile(df['Debt_Score'], 75, interpolation='midpoint')
q2 = df['Debt_Score'].median()
iqr = (q3-q1)
print('Q1  :',q1)
print('Q2  :',q2)
print('Q3  :',q3)

print('IQR :',iqr)
print('Upper Range for outlier :',q3+(1.5*iqr))
print('Lower Range for outlier :',q1-(1.5*iqr))
plt.grid()
plt.show()
```

```python
dbc = []
for i in df.Debt_Score:
    if i <= 0:
        dbc.append('Very Poor') ## < 0
    elif 1 <= i < q1 :
        dbc.append('Poor') ## 1 to 9 (q1)
    elif q1 <= i < q2 :
        dbc.append('Moderate') ## 9 (q1) to 17 (q2) 
    elif q2 <= i < q3 :
        dbc.append('Good') ## 17 (q2) to 30 (q3)
    elif i >= q3:
        dbc.append('Very Good') ## >than 30 (q3)
df['Debit_Score_Catg'] = dbc
```

```python
col = df.pop('Status')
df.insert(24,'Status',col)
```

```python
dbc_status = pd.DataFrame(df.Status.groupby(df.Debit_Score_Catg).value_counts())
dbc_status['Percentage'] = round(pd.DataFrame(df.Status.groupby(df.Debit_Score_Catg).value_counts(normalize=True))*100,2)
```

```python
dbc_status
```

### Inference : Now, we can clearly see the difference! The Debt score when less than 0 for a cust, will be surely denied and for customers who have debit scores greater than 0, will surely be given a CC.!
